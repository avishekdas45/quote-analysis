### Fetch all the Author,  Quotes, Tag Names from Website and create table

```python
def get_authors_quotes_tags():

    # Making a dictionary to store the page contents
    topic_dict={'Author':[], 'Quote':[], 'Tag_Names':[]}
    
    for page in range(1,11):
        topic_url= f'http://quotes.toscrape.com/page/{page}'
        response= requests.get(topic_url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text,'html.parser')
            
            # To fetch the Author names
            authors = soup.find_all('small', {'class':"author"})
            author_list = [a.text.strip() for a in authors]

            # To fetch the Quotes given by the Author
            quotes = soup.find_all('span', {'class':"text"})
            quote_list = [q.text.strip() for q in quotes]
            
            # To fetch the Tags of every Quote    
            tag_names = soup.find_all('div', {'class':"tags"})
            tag_list = []
            for t in tag_names:
                tags = [tag.text.strip() for tag in t.find_all('a', {'class':"tag"})]
                tag_list.append(','.join(tags))
            
            # Append data to dictionary
            topic_dict['Author'].extend(author_list)
            topic_dict['Quote'].extend(quote_list)
            topic_dict['Tag_Names'].extend(tag_list)


    return  topic_dict 
d= get_authors_quotes_tags()
```

```python
data=pd.DataFrame(d)
data
```

### Save the DataFrame

```python
data.to_csv('bw_project.csv', index = False)
print("Data has been saved to .csv format and file name is: 'bw_project.csv'")
```
